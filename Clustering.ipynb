```python
import pandas as pd

df = pd.read_csv('meritocracy_detailed3.csv')
filtered_df = df[df['time_step'] > 39]
df
```

```python
dfdum = pd.get_dummies(df['personality'], prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=float)
```

```python

from stepmix.stepmix import StepMix

# Categorical StepMix Model with 3 latent classes
model = StepMix(n_components=2, measurement="binary", verbose=1)
model.fit(dfdum)
```

```python

from sklearn.preprocessing import LabelEncoder


# Encode categorical variables
le = LabelEncoder()

# Fit StepMix model
model = StepMix(n_components=2, measurement="binary", verbose=1)
model.fit(dfdum)

# Save latent class assignments
latent_classes = model.predict(dfdum)
latent_df = dfdum.copy()
latent_df['latent_class'] = latent_classes

# Show the first few rows of the latent class assignments
display_df = latent_df.head()
print(display_df)
```

```python


latent_df['group_id'] = latent_df.index


# Merge datasets
merged_df = pd.merge(df, latent_df[['group_id', 'latent_class']],
                    on='group_id', how='left')

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np


pca_columns = ['promotion_time', 'group_score', 'group_share_normal',
               'group_share_overachiever', 'group_share_charismatic_idiot']

# Convert promotion_time to numeric, handling NaT values
df['promotion_time'] = pd.to_numeric(df['promotion_time'], errors='coerce')

# Remove rows with NaN values
df_clean = df.dropna(subset=pca_columns)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_clean[pca_columns])

pca = PCA()
pca_result = pca.fit_transform(scaled_data)

print('Explained Variance Ratio:')
print(pca.explained_variance_ratio_)

print('\
Cumulative Explained Variance Ratio:')
print(np.cumsum(pca.explained_variance_ratio_))

print('\
Rotation Matrix (Component Loadings):')
loadings_df = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(len(pca_columns))],
    index=pca_columns
)
print(loadings_df)
```

```python
# 3D Plot
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pca_result[:, 0], pca_result[:, 1], pca_result[:, 2],
                    c=df_clean['group_score'], cmap='viridis', alpha=0.5)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Plot')
plt.colorbar(scatter, label='Group Score')
plt.show()

print('Explained Variance Ratio:')
print(pca.explained_variance_ratio_)
```
