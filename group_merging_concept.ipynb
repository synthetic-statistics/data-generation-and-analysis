## Goal

Check all pairwise combinations of N samples in the "fairest" way possible, where "fair" means, that the waiting time between checks is minimal for each sample.

```python
def create_table(N):
    table={}
    
    shift=1
    first=True
    for j in range(N-1):    
        row=["-" for i in range(N)]
    
        pattern=np.array(dgf._shiftroll(shift,N))
        if first:
            extra="a"
        else:
            extra="b"
            end=N-len(pattern)
            pattern= (pattern+shift)[:end]
    
        for i in pattern:
            row[i]=(i+shift)%N
    
        table["shift "+str(shift)+" "+extra]=row
        
        if not first:
            shift+=1
        first=not first
    return table    
```

```python
N=16
print("Combination table for all pairwise checks")
print("in sequential order left to right and top to bottom")
pd.DataFrame.from_dict(create_table(N), orient='index',columns=np.arange(N))
```
